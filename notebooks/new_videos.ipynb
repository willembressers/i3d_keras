```python
from IPython.display import HTML
from base64 import b64encode
import os
```

```python
def show_input_video(video_name, video_format):
    mp4 = open('../data/input_videos/' + video_name +'.' + video_format,'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML("""
    <video width=400 controls>
        <source src="%s" type="video/mp4">
    </video>
    """ % data_url)

def preprocess_and_analyze_video(video_name, video_format):
    !rm ../data/*.npy
    !rm ../data/frames/*.jpg
    os.environ["NEW_VIDEO_PATH"] = "../data/input_videos/" + video_name + "." + video_format
    os.environ["VIDEO_TITLE"] = video_name
    !python ../src/preprocess.py --video_path ${NEW_VIDEO_PATH}
    !python ../src/evaluate_sample.py --video_name ${VIDEO_TITLE}
    !python ../src/visualize.py --video_name ${VIDEO_TITLE}
```
